**Review the article “Why is the Tower of God Show So Popular?” and imagine it has user datasuch as page views, average time spent, and bounce rate. Analyze these metrics and create visualizations (bar graphs, pie charts) to identify engagement trends.**

***

After going through the article, for better understanding we need to follow the below steps for further analysis.

**Step 1: Data Assumption**
-

As no actual user interaction data is provided, we'll assume the following metrices:

**Page views:** Number of users visiting the page. <br>
**Average Time Spent:** Average time spent by each user on the page. <br>
**Bounce Rate:** The percentage of users that leaves the page without further interaction. <br>
**Page Loading Time:** Time taken to fully load the page. <br>

Now, let's assume the following data points:

Page Views : 50000 <br>
Average Time Spent : 3 minutes <br>
Bounce Rate : 70% <br> 
Page Loading Time : 5 seconds <br>
Repeat Viewers : 25% <br>
New Viewers : 75% <br>

**Step 2: Data Analysis and Visualizations**
-

* ***Average Time Spent Analysis***

Average time spent on the page is 3 minutes which is very low. It suggests most of the users spend very less time to read articles on the page.

* ***Bounce Rate Analysis***

The bounce rate is 70% which is very high. It suggests that majority of the page views do not get further interaction on the page.

* ***Page Loading Time Analysis***

The page loading time is 5 seconds which is quite long. For a good quality website or server this value is 0-2 seconds.

* ***Analyzing Bounced Views from Page Views***

```python
page_views = 50000 
bounce_rate = 70 
bounced_views = int(page_views * bounce_rate / 100)

view_dist_labels = ['Page Views', 'Bounced Views']
view_dist_values = [page_views, bounced_views]

plt.bar(view_dist_labels, view_dist_values, edgecolor='black')
```

* ***Viewers Distribution Analysis (New Viewers Vs Repeat Viewers)***

```python
viewer_dist_label = ['Repeat Viewers', 'New Viewers']
viewer_dist_value = [25, 75]

explode = [0, 0.1]
plt.pie(viewer_dist_value, labels=viewer_dist_value, explode = explode, wedgeprops={'edgecolor' : 'black'})
plt.legend(viewer_dist_label, loc='upper right')
plt.title('Percentage of New Viewers Vs Repeat Viewers')
```

**Step 3: Insights**
-

* Majority (75%) of the users are new users and the rest (25%) are repeat users. It suggests that the content is reaching to a broad audience but may not be retaining them.

* Users find the content not much engaging as only 30% of total page views get further interaction, which is considered as very low. <br>
Longer page loading time is also a potential cause for high bounce rate. 

* The average time spent on the page is quite low. Its content does not engage users to spend more time on the page.
