**Step 2: Data Analysis and Visualizations**
-

* ***Daily Active Users Analysis***

The number of daily active users is around 300, which is quite low considering the total users.

* ***Average Time Spent Analysis***

Average time spent on the page is 3 minutes which is very low. It suggests most of the users spend very less time to read articles on the page.

* ***Bounce Rate Analysis***

The bounce rate is 70% which is very high. It suggests that majority of the page views do not get further interaction on the page.

* ***Page Loading Time Analysis***

The page loading time is 4 seconds which is longer than usual. For a good quality website or server this value is 0-2 seconds.

* ***Daily Active Users Analysis From Total Users***

```python
total_users = 30000
daily_users = 300 

users_labels = ['Total Users', 'Daily Active Users']
users_values = [total_users, daily_users]

plt.barh(users_labels, users_values, edgecolor = 'black')
```

* ***Analyzing Bounced Views from Page Views***

```python
page_views = 60000 
bounce_rate = 70 
bounced_views = int(page_views * bounce_rate / 100)

view_dist_labels = ['Page Views', 'Bounced Views']
view_dist_values = [page_views, bounced_views]

plt.bar(view_dist_labels, view_dist_values, edgecolor='black')
```

* ***Viewers Distribution Analysis (New Viewers Vs Repeat Viewers)***

```python
viewer_dist_label = ['Repeat Viewers', 'New Viewers']
viewer_dist_value = [30, 70]

explode = [0, 0.1]
plt.pie(viewer_dist_value, labels=viewer_dist_value, explode = explode, wedgeprops={'edgecolor' : 'black'})
plt.legend(viewer_dist_label, loc='upper right')
plt.title('Percentage of New Viewers Vs Repeat Viewers')
```
