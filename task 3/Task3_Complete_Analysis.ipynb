**Step 2: Customer Segment Data Analysis and Visualizations**
-

* ***New Users Vs Repeat Users Analysis***

```python
viewer_dist_label = ['Repeat Viewers', 'New Viewers']
viewer_dist_value = [30, 70]

explode = [0, 0.1]
plt.pie(viewer_dist_value, labels=viewer_dist_value, explode = explode, wedgeprops={'edgecolor' : 'black'})
plt.legend(viewer_dist_label, loc='upper right')
plt.title('Percentage of New Viewers Vs Repeat Viewers')
```

Majority (70%) of the users are New users, rest (30%) are the repeat users.

* ***Users by Gender Analysis***

```python
Gender_dist_label = ['Male Users', 'Female Users']
Gender_dist_value = [18000, 12000]

explode = [0, 0.1]
plt.pie(Gender_dist_value, labels=Gender_dist_value, explode = explode, wedgeprops={'edgecolor' : 'black'})
plt.legend(Gender_dist_label, loc='upper right')
plt.title('Percentage of Male Users Vs Female Users')
```
